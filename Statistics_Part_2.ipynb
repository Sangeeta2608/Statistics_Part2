#1.What is hypothesis testing in statistics?
Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data.It involves making an assumption and then using statistical methods to determine whether the data supports or refutes that assumption.


#2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
Null hypothesis is the default or starting assumption.It typically states that there is no effect, no difference, or no relationship between variables.
Example: H‚ÇÄ:The average height of students is 170 cm.
Alternative Hypothesis represents what are we trying to prove that there is an effect, difference, or relationship.
Example: H‚ÇÅ:The average height of students is not 170 cm.

#3.What is the significance level in hypothesis testing, and why is it important?
The significance level, often denoted by Œ±(alpha), is a threshold that you set before conducting a hypothesis test. It represents the maximum probability of making a Type I error‚Äîrejecting the null hypothesis (H‚ÇÄ) when it is actually true.

#4.What does a P-value represent in hypothesis testing?
The p-value in hypothesis testing represents the probability of obtaining a test result at least as extreme as the one observed, assuming that the null hypothesis(H‚ÇÄ) is true.



#5.How do you interpret the P-value in hypothesis testing?
The p-value helps us to decide whether to reject the null hypothesis(H‚ÇÄ) based on the evidence from your sample data.
The p-value is the probability of obtaining results as extreme as(or more extreme than) the observed results if H‚ÇÄ is true.
A small p-value (e.g., < 0.01) = strong evidence against H‚ÇÄ.
A large p-value (e.g., > 0.10) = weak or no evidence against H‚ÇÄ.
Failing to reject H‚ÇÄ ‚â† proving H‚ÇÄ is true. It only means you don't have enough evidence to say otherwise.







#6.What are Type 1 and Type 2 errors in hypothesis testing?
Type I Error(False Positive):Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.We say there's an effect or difference, but there isn't.
Probability of Type I error = Œ± (significance level).
Type II Error (False Negative):Failing to reject the null hypothesis (H‚ÇÄ) when it is actually false.We miss a real effect or difference.
Probability of Type II error = Œ≤.Power of a test = 1 - Œ≤ (probability of detecting a real effect).



#7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
The difference between a one-tailed and a two-tailed test in hypothesis testing lies in what kind of effect or difference you're testing for ‚Äî specifically, the direction of the effect.
one-tailed test:Tests for an effect in one specific direction either greater than or less than.
Two-tailed test:Tests for an effect in either direction ‚Äî different from the null, but not specifically greater or smaller.


#8. What is the Z-test, and when is it used in hypothesis testing?
A Z-test is a statistical test used to determine whether there is a significant difference between sample data and a population parameter (like the mean or proportion), under the assumption that the data follows a normal distribution and you know the population standard deviation (œÉ).
It is used when:1.The population standard deviation (œÉ) is known.
2.The sample size is large (typically n ‚â• 30) OR the population is normally distributed.
3.You are testing the mean or proportion.

#10.What is the T-distribution, and when should it be used instead of the normal distribution.
The t-distribution is a probability distribution used in hypothesis testing when the sample size is small and/or the population standard deviation (œÉ) is unknown.
It looks similar to the normal distribution (bell-shaped and symmetric), but it has heavier tails ‚Äî meaning it accounts for more variability when working with smaller samples.
The fatter tails of the t-distribution provide a more conservative estimate of statistical significance when your sample is small or your estimate of variability (s) is less reliable.
T-distribution  is used when:
1.Sample size is small (typically n < 30)
2.Population standard deviation (œÉ) is unknown
3.Data are approximately normally distributed


#11.What is the difference between a Z-test and a T-test?
Both the Z-test and the T-test are used to determine whether there's a significant difference between a sample statistic (like a mean) and a population parameter, or between two groups. The main difference lies in the assumptions about the population standard deviation and the sample size.
Z-test used when:
Population standard deviation (œÉ) is known.
Sample size is large (n ‚â• 30).
The data is approximately normally distributed (or the Central Limit Theorem applies).
T-test is used when:
Population standard deviation (œÉ) is unknown.
Sample size is small (n < 30).
The population is normally distributed, or the sample is approximately normal.

#12.What is the T-test, and how is it used in hypothesis testing?

Between a sample mean and a known value (one-sample t-test),
Two related samples (paired t-test), or
Two independent samples (independent t-test).
It‚Äôs most commonly used when the population standard deviation (œÉ) is unknown and the sample size is small (typically n < 30).
A T-test is used when:
1.we don't know œÉ (the population standard deviation),
2.working with small samples,
3.the data is approximately normally distributed.



#13.What is the relationship between Z-test and T-test in hypothesis testing?
The Z-test and T-test are both used to assess whether a sample mean (or difference between means) is statistically different from a known value or another sample mean. They are conceptually very similar ‚Äî the main difference lies in assumptions about the data and sample characteristics.



#14. What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval (CI) is a range of values used to estimate an unknown population parameter (such as the mean or proportion) based on sample data. It provides not just a single estimate, but a range that likely contains the true value.



#15.What is the margin of error, and how does it affect the confidence interval?
The margin of error is the maximum amount you expect your sample estimate to differ from the true population value, given a certain level of confidence. It‚Äôs a critical part of constructing and interpreting a confidence interval.
Wider margin of error ‚Üí wider confidence interval ‚Üí less precise estimate.
Smaller margin of error ‚Üí narrower confidence interval ‚Üí more precise estimate.

#17. What is the Chi-square distribution, and when is it used?
The Chi-square distribution is a right-skewed, continuous probability distribution that arises in statistics when you're dealing with sums of squared values, especially in categorical data analysis.
It is used mainly for testing relationships between categorical variables and evaluating variance.

#18.What is the Chi-square goodness of fit test, and how is it applied?

#19.What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is a continuous, right-skewed probability distribution that arises when comparing two variances or testing multiple means. It is mainly used in ANOVA (Analysis of Variance) and in comparing variances between two or more samples.



#20.What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) is a statistical test used to determine if there are statistically significant differences between the means of three or more groups. It helps answer the question:‚ÄúAre the group means equal or do at least one of them differ significantly?‚Äù
The results of an ANOVA to be valid, certain assumptions must be met:
1. Independence of Observations
Each observation in the data should be independent of the others. This means that the values in one group should not influence the values in another group.
2. Normality
The data in each group should be approximately normally distributed. This assumption can be checked using a normality test (e.g., Shapiro-Wilk test) or by visualizing the data with histograms or Q-Q plots.
3. Homogeneity of Variances (Homoscedasticity)
The variance within each group should be approximately equal. This assumption can be tested using Levene's test or Bartlett‚Äôs test. If the variances are not equal, it might indicate that ANOVA is not appropriate, or a more robust method (like Welch‚Äôs ANOVA) should be used.

#21.What are the different types of ANOVA tests?
Types of ANOVA Test:
1.One way Anova
2.Two way Anova
3.Repeated Measures ANOVA
4.Manova
5.Welch‚Äôs ANOVA
6.MixedDesign Anova

#22. What is the F-test, and how does it relate to hypothesis testing?
The F-test is a statistical test used to compare two variances or to test the overall significance of a model, particularly in regression analysis and ANOVA (Analysis of Variance). It assesses whether the variances between two or more groups are significantly different or whether a model is a good fit for the data.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else (1 - p)
print(bernoulli_pdf(1, 0.7))  # Output: 0.7
print(bernoulli_pdf(0, 0.7))  # Output: 0.3
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.scatter(range(len(z_scores)), z_scores, color='blue', alpha=0.6)
    plt.axhline(0, color='red', linestyle='--')
    plt.title("Z-Scores (Scatter)")
    plt.xlabel("Data Point Index")
    plt.ylabel("Z-Score")
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=30, color='green', edgecolor='black', alpha=0.7)
    plt.title("Z-Scores (Histogram)")
    plt.xlabel("Z-Score")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

    return z_scores
data = np.random.normal(loc=50, scale=10, size=100)
z_scores = calculate_and_plot_z_scores(data)
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

def demonstrate_clt(sample_size=30, num_samples=1000):
    original_data = np.random.exponential(scale=2.0, size=10000)
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.choice(original_data, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=50, color='orange', edgecolor='black', alpha=0.7)
    plt.title('Original Exponential Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=40, color='skyblue', edgecolor='black', alpha=0.7, density=True)
    plt.title(f'Distribution of Sample Means (n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

    plt.tight_layout()
    plt.show()
demonstrate_clt(sample_size=30, num_samples=1000)
```

```python
#11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def verify_clt_with_normal_dist(sample_size=30, num_samples=1000, mean=0, std_dev=1):
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
        sample_means.append(np.mean(sample))
    plt.figure(figsize=(10, 6))
    plt.hist(sample_means, bins=40, color='lightgreen', edgecolor='black', density=True, alpha=0.7)
    x = np.linspace(min(sample_means), max(sample_means), 100)
    sample_mean = mean
    sample_std = std_dev / np.sqrt(sample_size)
    plt.plot(x, norm.pdf(x, loc=sample_mean, scale=sample_std), 'r-', lw=2, label='Theoretical Normal PDF')
    plt.title(f'Distribution of Sample Means (CLT Verification)\nSample Size = {sample_size}, Num Samples = {num_samples}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True)
    plt.legend()
    plt.show()
verify_clt_with_normal_dist(sample_size=30, num_samples=1000, mean=0, std_dev=1)
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (x - mean) / std_dev
    cdf_value = norm.cdf(z_score)
    print(f"Z-score for {x}: {z_score:.4f}")
    print(f"Cumulative Probability (CDF) for Z-score {z_score:.4f}: {cdf_value:.4f}")

    return z_score

def plot_standard_normal_distribution(z_score):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)
    plt.plot(x, y, label='Standard Normal Distribution', color='blue', lw=2)
    plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.fill_between(x, y, where=(x >= z_score), color='red', alpha=0.3)
    plt.title(f'Standard Normal Distribution with Z-score = {z_score:.2f}')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
data = np.random.normal(loc=50, scale=10, size=1000)
x = 55

z_score = calculate_z_score(data, x)
plot_standard_normal_distribution(z_score)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(np.abs(z_statistic)))
    reject_null = p_value < alpha
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Reject null hypothesis: {'Yes' if reject_null else 'No'}")

    return z_statistic, p_value, reject_null
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)
population_mean = 52
population_std = 10
z_statistic, p_value, reject_null = z_test(sample_data, population_mean, population_std, alpha=0.05)
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    data = np.array(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Sample Size: {sample_size}")
    print(f"Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

    return lower_bound, upper_bound
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)
lower_bound, upper_bound = confidence_interval(sample_data, confidence_level=0.95)
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy import stats
def z_test(sample, population_mean, population_std):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    return z_statistic, p_value
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=100)
population_mean = 50
population_std = 10
z_stat, p_value = z_test(sample, population_mean, population_std)
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05, test_type='right'):
    se = sigma / np.sqrt(n)
    if test_type == 'right':
        z_crit = norm.ppf(1 - alpha)
        x_crit = mu0 + z_crit * se
    elif test_type == 'left':
        z_crit = norm.ppf(alpha)
        x_crit = mu0 + z_crit * se
    elif test_type == 'two-tailed':
        z_crit_right = norm.ppf(1 - alpha / 2)
        z_crit_left = norm.ppf(alpha / 2)
        x_crit_right = mu0 + z_crit_right * se
        x_crit_left = mu0 + z_crit_left * se
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='Null Hypothesis $H_0$', color='blue')
    plt.plot(x, y1, label='Alternative Hypothesis $H_1$', color='green')

    if test_type == 'right':
        # Type I Error: area to the right of critical value under H0
        plt.fill_between(x, y0, where=(x >= x_crit), color='red', alpha=0.3, label='Type I Error (Œ±)')
        # Type II Error: area to the left of critical value under H1
        plt.fill_between(x, y1, where=(x < x_crit), color='orange', alpha=0.3, label='Type II Error (Œ≤)')
        plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')

    elif test_type == 'left':
        plt.fill_between(x, y0, where=(x <= x_crit), color='red', alpha=0.3, label='Type I Error (Œ±)')
        plt.fill_between(x, y1, where=(x > x_crit), color='orange', alpha=0.3, label='Type II Error (Œ≤)')
        plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')

    elif test_type == 'two-tailed':
        plt.fill_between(x, y0, where=(x <= x_crit_left) | (x >= x_crit_right), color='red', alpha=0.3, label='Type I Error (Œ±)')
        plt.fill_between(x, y1, where=(x > x_crit_left) & (x < x_crit_right), color='orange', alpha=0.3, label='Type II Error (Œ≤)')
        plt.axvline(x_crit_left, color='black', linestyle='--', label=f'Left Critical = {x_crit_left:.2f}')
        plt.axvline(x_crit_right, color='black', linestyle='--', label=f'Right Critical = {x_crit_right:.2f}')
    plt.title('Type I and Type II Error Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=36, alpha=0.05, test_type='right')
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return lower, upper

np.random.seed(0)
sample = np.random.normal(loc=100, scale=15, size=25)

ci = confidence_interval(sample, confidence=0.95)
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    std_dev = np.std(data, ddof=1)
    std_err = std_dev / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_crit * std_err
    return moe

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): ¬±{moe:.2f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
from scipy.stats import binom

def bayesian_inference(prior_A, prior_B, flips, heads):
    likelihood_A = binom.pmf(heads, flips, 0.5)
    likelihood_B = binom.pmf(heads, flips, 0.8)
    evidence = (likelihood_A * prior_A) + (likelihood_B * prior_B)
    posterior_A = (likelihood_A * prior_A) / evidence
    posterior_B = (likelihood_B * prior_B) / evidence

    return {
        'Posterior_A': posterior_A,
        'Posterior_B': posterior_B
    }
prior_A = 0.5
prior_B = 0.5
flips = 5
heads = 4
posteriors = bayesian_inference(prior_A, prior_B, flips, heads)
print("Posterior probabilities after observing 4 heads in 5 flips:")
print(f"Coin A (fair):  {posteriors['Posterior_A']:.4f}")
print(f"Coin B (biased): {posteriors['Posterior_B']:.4f}")

```

```python
#13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    if not isinstance(observed, pd.DataFrame):
        raise ValueError("Input must be a pandas DataFrame for labeling clarity.")
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.values.sum()
    expected = np.outer(row_totals, col_totals) / grand_total
    return pd.DataFrame(expected, index=observed.index, columns=observed.columns)
data = pd.DataFrame({
    'Coffee': [30, 20],
    'Tea': [10, 40]
}, index=['Young', 'Old'])

print("Observed Frequencies:")
print(data)

expected = calculate_expected_frequencies(data)

print("\nExpected Frequencies (assuming independence):")
print(expected.round(2))
```
